```python
Population_by_state = df.groupby('STATE')['ESTIMATED_POP'].sum().reset_index()
Population_by_state
```

```python
population_by_city = df.groupby('CITY')['ESTIMATED_POP'].sum().reset_index()
population_by_city
```

```python
# Sort the DataFrame by population in descending order
Population_by_state10 = Population_by_state.sort_values(by=['ESTIMATED_POP'], ascending=False)

# Convert the population values to integers
Population_by_state10['ESTIMATED_POP'] = Population_by_state10['ESTIMATED_POP'].astype(int)

# Reset the index of the DataFrame
Population_by_state10 = Population_by_state10.reset_index()

# Select the top 10 rows
Population_by_state10 = Population_by_state10.head(10)

# Drop the 'index' column
Population_by_state10.drop('index', axis=1, inplace=True)

# Display the resulting DataFrame
print(Population_by_state10)
```

```python
Population_by_state['STATE'].value_counts().sum()
```

```python
population_by_city10 = population_by_city.sort_values(by=['ESTIMATED_POP'], ascending=False)
population_by_city10['ESTIMATED_POP'] = population_by_city10['ESTIMATED_POP'].astype(int)
population_by_city10 = population_by_city10.reset_index()
population_by_city10 = population_by_city10.head(10)
population_by_city10
```

```python
population_by_city10
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10,5))
plt.ticklabel_format(style='plain', axis='y')
ax = sns.barplot(x ='STATE', y ='ESTIMATED_POP', data = Population_by_state10, palette ='viridis')
ax.bar_label(ax.containers[0],fmt='%.0f')
plt.title('Top 10 populations by states')
plt.show() 
```

```python
sns.set_style('darkgrid')
plt.figure(figsize=(12,5))
plt.ticklabel_format(style='plain', axis='y')
ax = sns.barplot(x ='CITY', y ='ESTIMATED_POP', data = population_by_city10, palette ='magma')
ax.bar_label(ax.containers[0],fmt='%.0f')
plt.title('Top 10 populations by cities')
plt.show()
```
